# Método de Euler según
# [Python Programming And Numerical Methods: A Guide For Engineers And Scientists](https://pythonnumericalmethods.berkeley.edu/notebooks/chapter22.03-The-Euler-Method.html)

```python
# import numpy as np
# import matplotlib.pyplot as plt

# plt.style.use('seaborn-poster')
# %matplotlib inlinea

# Parámetros
f = lambda x, y: np.exp(-x)  # EDO
h = 0.1  # Tamaño del paso
x = np.arange(0, 4 + h, h) # Numerical grid
y0 = -1  # Condición inicial

dydx = -np.exp(-x)

# Método de Euler explícito
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h*f(x[i], y[i])

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.plot(x, dydx, 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

## Función del libro **Métodos Numéricos Para Ingenieros**
## 25.1 MÉTODO DE EULER

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Parámetros
f = lambda x, y: -0.5 * x**4 + 4 * x**3 - 10 * x**2 + 8.5 * x + 1  # EDO
h = 0.5  # Tamaño del paso
x = np.arange(0, 4 + h, h) # Numerical grid
y0 = 1  # Condición inicial

dydx = -2 * x**3 + 12 * x**2 - 20 * x + 8.5

# Método de Euler explícito
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h*f(x[i], y[i])

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.plot(x, dydx, 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

## Función $f(x) = x^2$
## Derivada $f'(x) = 2x$

```python
# import numpy as np
# import matplotlib.pyplot as plt

# plt.style.use('seaborn-poster')
# %matplotlib inlinea

# Parámetros
f = lambda x, y: x**2  # EDO
h = 0.1  # Tamaño del paso
x = np.arange(0, 4 + h, h) # Numerical grid
y0 = -1  # Condición inicial

dydx = 2 * x

# Método de Euler explícito
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h*f(x[i], y[i])

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.plot(x, dydx, 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
